```python
columns_rename={'ts':'timestamp','device':'device_id','co':'carbon_monoxide','humidity':'humidity'
                ,'light':'light','lpg':'liquefied_petroleum_gas','motion':'motion','smoke':'smoke','temp':'temperature'}
```

```python
df=df.rename(columns=columns_rename)
```

```python
df
```

```python
df=df.isnull().sum()       # find null values
df
```

```python
df=df.drop_duplicates()         # drop duplicates
```

```python
df=df.dropna()           # drop null values
```

```python
df.dtypes # dtypes
```

```python
df['timestamp'] = df['timestamp'].astype('datetime64[ns]')   # data type changing
```

```python
df['light']=df['light'].astype('bool')
```

```python
df['motion']=df['motion'].astype('bool')
```

```python
df=df.to_csv("cleaned_dataset.csv")      # load the csv file
```
